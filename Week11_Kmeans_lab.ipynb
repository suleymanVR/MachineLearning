```python
#@title Varsayılan başlık metni
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# %% create dataset

# class1
x1 = np.random.normal(25,5,1000)
y1 = np.random.normal(25,5,1000)

# class2
x2 = np.random.normal(55,5,1000)
y2 = np.random.normal(60,5,1000)

# class3
x3 = np.random.normal(55,5,1000)
y3 = np.random.normal(15,5,1000)

# class4
x4 = np.random.normal(10,5,1000)
y4 = np.random.normal(65,5,1000)

# class5
x5 = np.random.normal(30,5,1000)
y5 = np.random.normal(65,5,1000)
```

```python
x = np.concatenate((x1,x2,x3),axis = 0)
y = np.concatenate((y1,y2,y3),axis = 0)
```

```python
dictionary = {"x":x,"y":y}

data = pd.DataFrame(dictionary)

plt.scatter(x1,y1,color = "blue")
plt.scatter(x2,y2,color = "red")
plt.scatter(x3,y3,color = "green")
plt.scatter(x4,y4,color = "black")
plt.scatter(x5,y5,color = "orange")
plt.show()
```

```python
# %% KMEANS

from sklearn.cluster import KMeans

#Within Clusters Sum of Square (WCSS) (Kümeler içi kareler toplamı)
wcss = [] 

for k in range(1,15):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,15),wcss)
plt.xlabel("number of k (cluster) value")
plt.ylabel("Within Clusters Sum of Square(wcss)")
plt.show()
```

```python
#Ustteki görselde kırılma noktasından sonraki noktada 3 oldugu icin 3 yeterli olacaktır.
#%% k = 3 icin modelim

for i in range(2,6):
  kmeans2 = KMeans(n_clusters=i)
  clusters = kmeans2.fit_predict(data)

  data["label"] = clusters

  plt.scatter(data.x[data.label == 0 ],data.y[data.label == 0],color = "red")
  plt.scatter(data.x[data.label == 1 ],data.y[data.label == 1],color = "green")
  plt.scatter(data.x[data.label == 2 ],data.y[data.label == 2],color = "blue")
  plt.scatter(data.x[data.label == 3 ],data.y[data.label == 3],color = "black")
  plt.scatter(data.x[data.label == 4 ],data.y[data.label == 4],color = "orange")
  plt.scatter(kmeans2.cluster_centers_[:,0],kmeans2.cluster_centers_[:,1],color = "yellow")
  plt.show()

  #color ="" yazılmadıgında random renk ataması yapar.
```
